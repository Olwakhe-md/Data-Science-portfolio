```python
def read_pdf_file(pdf_path):
    print("reading the pdf file")
    text = " "
    with pdfplumber.open(pdf_path) as pdf:
        for page_number, page in enumerate(pdf.pages, 1):#iterate over all the pages to know the number of pages its reading
            print(f"   Reading page {page_number}...")
            page_text = page.extract_text()
            if page_text:  
                text += page_text + "\n"
    print(f"✅ Finished reading PDF. Total characters: {len(text)}")
    return text
```

```python
def find_plant_names(text):
    print("🔍 Looking for plant names...")
    plants = [] #create a new list to store your plants
    pattern = r'([A-Z\s]+)\s*\(([A-Za-z\s]+)[^)]*\)\s*\++' #study the pattern of the plant names
    matches = re.findall(pattern, text) #find all the names that match the pattern etc
    for common_name, scientific_name in matches:
        plant_info = {
            'common_name': common_name.strip(),
            'scientific_name': scientific_name.strip() #iterate and clean the names
        }
        plants.append(plant_info) # add the found plant names to the list created in the beginning 
        print(f"   Found: {plant_info['common_name']} ({plant_info['scientific_name']})")
    
    print(f"✅ Found {len(plants)} plants total") # tells you how many plants you have managed to find
    return plants
```

```python
def clean_item(item):
        # Remove parenthesis contents and trailing punctuation
        item = re.sub(r'\(.*?\)', '', item)  # remove (....)
        item = re.sub(r'[^a-zA-Z\s\-]', '', item)
        item = item.replace('[', '').replace(']', '').replace("'", "")
        item = item.replace('\n', '')
        item = item.strip()                  # trim spaces
        return item
cleaned = [clean_item(i) for i in items if clean_item(i)]
cleaned
```
